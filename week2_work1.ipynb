### 2.二次元配列と勾配問題
機械学習においては関数の最小値を見つけることが重要になってきます。  

そのために使われるのが勾配降下法です。  
この課題では勾配降下法を理解する準備として、  
NumPyを用いて関数の勾配を考えていきます。

####  【問題1】1次関数
 xが-50から50の範囲で次の式を考えます。  
   
   $ y = \frac{1}{2}x + 1 $  
     
 -50から50まで0.1間隔のxを表すndarrayと,  
 各要素に対するyを表すndarrayを作成してください。

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(-50,50.1,0.1).reshape(-1,1)
y = np.array(x/2 + 1)

print('x shape :',x.shape)
print('y shape :',y.shape)
print('x ndarray :\n',x)
print()
print('y ndarray :\n',y)
```

#### 【問題2】配列の結合
問題1で作成したndarrayを結合し、(1001, 2)の形の配列にしてください。  
0列目にx、1列目にyが来るようにします。

```python
xy = np.concatenate([x,y],1)
print('xy shape :',xy.shape)
print('xy ndarray :\n',xy)
```

#### 【問題3】勾配を求める
各x座標における勾配を求めます。  
ここで勾配はyの変化量をxの変化量で割ることで計算します。
  
《注意》

・厳密に各点における勾配を求めるには極限（微分）を利用する必要があります  
・勾配はx座標の間隔を使うので、xの要素数に対して一つ少なくなります

```python
gradient = ((xy[1:,1] - xy[:-1,1])/(xy[1:,0] - xy[:-1,0])).reshape(-1,1)
print('gradient shape :',gradient.shape)
print('gradient ndarray :\n',gradient)
```

#### 【問題4】グラフを描く
1次関数のグラフをMatplotlibを用いて描いてください。  
また、それとは別に横軸をx、縦軸を勾配としたグラフも描いてください。

```python
plt.title('y - x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(x,y,color = 'k')
plt.show()
```

```python
plt.title('gradient - x')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(x[:len(gradient)],gradient,color = 'k')
plt.show()
```

#### 【問題5】Pythonの関数化
問題1から問題3までの範囲をまとめて行うPythonの関数を作成してください。  
そして、以下の3式の勾配を0.1間隔で求めてndarrayで出力してください。 
  
$ y = x^{2} \quad (-50\leq x \leq 50)\\
y = 2x^{2}+2^{x} \quad (-50\leq x \leq 50)\\
y = sin(x^{\frac{1}{2}}) \quad (0\leq x \leq 50) $  
また、問題4と同様にグラフを描いてください。  
ndarrayに対して三角関数のsinの計算を行うには np.sin()を用います。

```python
def x_ndarray(x1 = -50, x2 = 50.1, d = 0.1):
    
    x = np.arange(x1,x2,d).reshape(-1,1)
    
    return x

    
def gradient_calcuraiton(y,x):
    

    xy = np.concatenate([x,y],1)
    gradient = ((xy[1:,1] - xy[:-1,1])/(xy[1:,0] - xy[:-1,0])).reshape(-1,1)
    

    return xy, gradient
```

$ y = x^{2} \quad (-50\leq x \leq 50)$

```python
# y = x**2
xy,gradient = gradient_calcuraiton(x_ndarray()**2,
                                   x_ndarray())
print('xy : ndarray\n',xy)
print('gradient :\n',gradient)
```

```python
plt.title('y - x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xy[:,0],xy[:,1],color = 'k')
plt.show()
```

```python
plt.title('gradient - x')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(xy[:len(gradient),0],gradient,color = 'k')
plt.show()
```

$ y = 2x^{2}+2^{x} \quad (-50\leq x \leq 50) $  

```python
# y = 2*x**2 + 2**x
xy,gradient = gradient_calcuraiton(2*x_ndarray()**2 + 2**x_ndarray(),
                                   x_ndarray())
print('xy : ndarray\n',xy)
print('gradient :\n',gradient)
```

```python
plt.title('y - x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xy[:,0],xy[:,1],color = 'k')
plt.show()
```

```python
plt.title('gradient - x')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(xy[:len(gradient),0],gradient,color = 'k')
plt.show()
```

$ y = sin(x^{\frac{1}{2}}) \quad (0\leq x \leq 50) $  

```python
# y = sin(x**0.5), 0 <= x <= 50 
xy,gradient = gradient_calcuraiton(np.sin(x_ndarray(0,50.1,0.1)**0.5),
                                          x_ndarray(0,50.1,0.1))
print('xy : ndarray\n',xy)
print('gradient :\n',gradient)
```

```python
plt.title('y - x')
plt.xlabel('x')
plt.ylabel('y')
plt.plot(xy[:,0],xy[:,1],color = 'k')
plt.show()
```

```python
plt.title('gradient - x')
plt.xlabel('x')
plt.ylabel('gradient')
plt.plot(xy[:len(gradient),0],gradient,color = 'k')
plt.show()
```

#### 【問題6】最小値を求める
問題5で扱った関数のyの最小値を求めます。xの範囲も問題5と同様とします。  
今回は勾配降下法ではなく、ndarrayのminメソッド、argminメソッドを用いてください。minメソッドは最小値を、argminメソッドは最小値のインデックスを取得することができます。

$ y = x^{2} \quad (-50\leq x \leq 50)$

```python
# y = x**2
xy,gradient = gradient_calcuraiton(x_ndarray()**2,
                                   x_ndarray())

miny = xy[:,1].min()
miny_index = xy[:,1].argmin()

print('最小のy : {:.1f}'.format(miny))
print('最小のyとなるインデックス番号 : {}'.format(miny_index))
```

```python
print('最小yの前後20個の勾配 : \n{}'
      .format(gradient[miny_index -20:miny_index +20]))
```

```python
# y = 2*x**2 + 2**x
xy,gradient = gradient_calcuraiton(2*x_ndarray()**2 + 2**x_ndarray(),
                                   x_ndarray())

miny = xy[:,1].min()
miny_index = xy[:,1].argmin()

print('最小のy : {:.1f}'.format(miny))
print('最小のyとなるインデックス番号 : {}'.format(miny_index))
```

```python
print('最小yの前後20個の勾配 : \n{}'
      .format(gradient[miny_index -20:miny_index +20]))
```

```python
# y = sin(x**0.5), 0 <= x <= 50 
xy,gradient = gradient_calcuraiton(np.sin(x_ndarray(0,50.1,0.1)**0.5),
                                          x_ndarray(0,50.1,0.1))

miny = xy[:,1].min()
miny_index = xy[:,1].argmin()

print('最小のy : {:.1f}'.format(miny))
print('最小のyとなるインデックス番号 : {}'.format(miny_index))
```

```python
print('最小yの前後20個の勾配 : \n{}'
      .format(gradient[miny_index -20:miny_index +20]))
```

メモ  
・関数の引数としてyとxを渡す時は、xとyを定義してから関数に渡すこと。  
　xを関数内で定義していたため、関数外で定義していたyの要素数と異なりエラーとなった
 （途中まで動いていたのは古いxで作成していたため）
