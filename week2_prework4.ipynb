```python
import numpy as np
import matplotlib.path as mpath
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
np.random.seed(0)
dice = np.array([1,2,3,4,5,6])
count = 1000
taget_num = np.random.choice(dice,count)

mean = [0,0]
cov = [[0.01,0],
       [0,0.01]]
darts = np.random.multivariate_normal(mean,cov,count)

result = np.insert(darts,0,taget_num,axis = 1)
print(result)
```

```python
dic = {}

for i in result:
    if i[0] not in dic:
        dic[i[0]] = [[i[1],i[2]]]
    else:
        dic[i[0]].append([i[1],i[2]])
        
darts_1 = np.array(dic[1])
darts_2 = np.array(dic[2])
darts_3 = np.array(dic[3])
darts_4 = np.array(dic[4])
darts_5 = np.array(dic[5])
darts_6 = np.array(dic[6])
```

```python
fig = plt.figure(figsize=(18, 3), dpi=600)
ax = plt.axes()

ax.scatter(1+darts_1[:,0], darts_1[:,1],color = 'blue'  ,alpha = 0.3, zorder = 1)
ax.scatter(2+darts_2[:,0], darts_2[:,1],color = 'red'   ,alpha = 0.3, zorder = 1)
ax.scatter(3+darts_3[:,0], darts_3[:,1],color = 'brown' ,alpha = 0.3, zorder = 1)
ax.scatter(4+darts_4[:,0], darts_4[:,1],color = 'green' ,alpha = 0.3, zorder = 1)
ax.scatter(5+darts_5[:,0], darts_5[:,1],color = 'orange',alpha = 0.3, zorder = 1)
ax.scatter(6+darts_6[:,0], darts_6[:,1],color = 'purple',alpha = 0.3, zorder = 1)

e1 = patches.Ellipse(xy=(1, 0), width=1, height=1, fc='w', ec='b',zorder = 0)
e2 = patches.Ellipse(xy=(2, 0), width=1, height=1, fc='w', ec='b',zorder = 0)
e3 = patches.Ellipse(xy=(3, 0), width=1, height=1, fc='w', ec='b',zorder = 0)
e4 = patches.Ellipse(xy=(4, 0), width=1, height=1, fc='w', ec='b',zorder = 0)
e5 = patches.Ellipse(xy=(5, 0), width=1, height=1, fc='w', ec='b',zorder = 0)
e6 = patches.Ellipse(xy=(6, 0), width=1, height=1, fc='w', ec='b',zorder = 0)

f1 = patches.Ellipse(xy=(1, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)
f2 = patches.Ellipse(xy=(2, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)
f3 = patches.Ellipse(xy=(3, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)
f4 = patches.Ellipse(xy=(4, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)
f5 = patches.Ellipse(xy=(5, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)
f6 = patches.Ellipse(xy=(6, 0), width=0.6, height=0.6, fc='w', ec='b',zorder = 0)

g1 = patches.Ellipse(xy=(1, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)
g2 = patches.Ellipse(xy=(2, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)
g3 = patches.Ellipse(xy=(3, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)
g4 = patches.Ellipse(xy=(4, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)
g5 = patches.Ellipse(xy=(5, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)
g6 = patches.Ellipse(xy=(6, 0), width=0.3, height=0.3, fc='w', ec='b',zorder = 0)

ax.add_patch(e1)
ax.add_patch(e2)
ax.add_patch(e3)
ax.add_patch(e4)
ax.add_patch(e5)
ax.add_patch(e6)

ax.add_patch(f1)
ax.add_patch(f2)
ax.add_patch(f3)
ax.add_patch(f4)
ax.add_patch(f5)
ax.add_patch(f6)

ax.add_patch(g1)
ax.add_patch(g2)
ax.add_patch(g3)
ax.add_patch(g4)
ax.add_patch(g5)
ax.add_patch(g6)
```

```python
def make_circle(r):
    t = np.arange(0, np.pi * 2.0, 0.01)
    t = t.reshape((len(t), 1))
    x = r * np.cos(t)
    y = r * np.sin(t)
    
    return np.hstack((x, y))

circle_1 = make_circle(0.1)
circle_2 = make_circle(0.3)
circle_3 = make_circle(0.5)


vertices = np.concatenate((circle_1,
                           circle_2,
                           circle_3))
plt.scatter(vertices[:,0],vertices[:,1])
plt.scatter(darts[:,0],darts[:,1])
```
