```python
import numpy as np
import pandas as pd
import matplotlib.path as mpath
import matplotlib.pyplot as plt
```

```python
def throw_darts(dice_num = 6,try_number=1000):

    np.random.seed(0)
    dice = np.arange(1,dice_num+1)

    target_number = np.random.choice(dice,try_number)

    mean = [0,0]
    cov = [[0.01,0],
           [0,0.01]]

    darts = np.random.multivariate_normal(mean,cov,try_number)

    result = pd.DataFrame({'target':target_number,'dart_x':darts[:,0],'dart_y':darts[:,1]})
       
    return dice,result

def make_circle(r=0.1,s=0):
    
    t = np.arange(0, np.pi * 2.0, 0.01)
    t = t.reshape(-1,1)
    
    x = (r*np.cos(t)).reshape(-1,1)
    y = (r*np.sin(t)).reshape(-1,1)
    
    shift_x = np.zeros(len(t)*len(r)).reshape(-1,1) + s 
    shift_y = np.zeros(len(t)*len(r)).reshape(-1,1) * s

    xy = np.concatenate([(x+shift_x).reshape(-1,1),(y+shift_y).reshape(-1,1)],axis=1)
    
    return xy


fig = plt.figure(figsize=(18, 3), dpi=200)
ax = plt.axes()

dice,result = throw_darts(12,2000)

radius = [0.1,
          0.3,
          0.5]

circle = make_circle(radius,dice)

plt.scatter(circle[:,0],circle[:,1],color='b',s=1)

cmap = plt.get_cmap('tab20')

plt.scatter(result['target']+result['dart_x'],
            result['dart_y'],
            alpha = 0.3,
            color = cmap(result['target']))
plt.show()
```
