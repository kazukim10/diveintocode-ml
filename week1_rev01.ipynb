【問題2】単位の変換

```python
t = 0.00008
t = t*(2**43)
print("厚さ： {}メートル".format(t))
print("厚さ： {:.2f}万キロメートル".format(t/10000000))
```

【問題3】for文を使用して作成

```python
t = 0.00008 #unit = m

for i in range(43):
    t = t*2
print("厚さ： {:.2f}万キロメートル".format(t/10000000))
```

【問題4】計算時間の比較

```python
t = 1
n = 10000

import time

start = time.time()
t = t*(2**n)
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))

t = 1
n = 10000

start = time.time()
for i in range(n):
    t = t*2
elapsed_time = time.time() - start
print("time : {}[s]".format(elapsed_time))
```

繰り返し回数が100程度であれば、べき乗の計算もfor文による計算も差はあまり無い。
ただし、計算量が10000になってくると差が顕著になる。（べき乗の計算速度＜for分の計算速度）

for文の計算量　1秒間で処理できる for 文ループの回数は、10^8=100,000,000回程度


【問題5】リストへの保存

```python
n = 43
#####
t = 0.00008 #unit = m
y = [t]

for i in range(n):
    t = t*2
    y.append(t)
    
print("リストのデータ数は",len(y))
print("厚さ： {:.2f}万キロメートル".format(t/10000000))
#####
```

【問題6】折れ線グラフの表示

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(y) # 「リスト名」のところにリストの変数名を入れる
plt.show()
```
