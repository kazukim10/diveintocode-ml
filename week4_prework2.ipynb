```python
#  ライブラリーをインポート
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# House Pricesのtrainデータを読み込む
train = pd.read_csv('./house-prices-advanced-regression-techniques/train.csv')
print(train.head())
print('shape:',train.shape)
```

## 回帰問題の流れ
回帰問題の流れは分類問題と基本的には同じですから、課題1のように細かい説明は行いません。

### 練習のための準備
このデータセットには79種類の特徴量（目的変数）がありますが、  
Week3で探索して分かったように様々なタイプのデータ含まれています。  
  
これをどう前処理するかということは機械学習において重要な箇所ですが、  
今回は回帰問題の学習や評価の練習を優先するため、  
説明変数(特徴量)GrLivAreaとYearBuiltの2つだけを利用することにします。

### 【問題1】練習のために特徴量の選択

```python
# 説明変数と目的変数の抽出
x = train[['GrLivArea','YearBuilt']]
y = train[['SalePrice']]
print('x:\n',x.head())
print('y:\n',y.head())
```

### 【問題2】線形回帰による推定と評価
* データを簡単に分析、前処理し、線形回帰 による  
  学習、推定、評価、可視化までの一連の流れを作成・実行してください。

sklearn.linear_model.LinearRegression — scikit-learn 0.21.3 documentation
  https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html
  
* 評価のための指標値は回帰問題では平均二乗誤差（Mean Squared Error, MSE）を使うことが一般的です。
  
sklearn.metrics.mean_squared_error — scikit-learn 0.21.3 documentation
  https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_squared_error.html
  
可視化は横軸にある特徴量、縦軸に推測した住宅価格といった方法で行えます。  
scatterを用いて、実際のデータと予測結果を比較できるグラフを作りましょう。

```python
# 欠損値の確認
print(x.isnull().sum())
print(y.isnull().sum())
```

```python
# データの最大、最小、平均、四分位などを確認
print(x.describe())
print(y.describe())
```

```python
# 簡単な散布図で、データ間の関係性を確認

fig,ax = plt.subplots(1,2,sharey=True,figsize=(16, 5))
ax[0].scatter(x['GrLivArea'],y)
ax[1].scatter(x['YearBuilt'],y);
```

* <GrLivArea>単位：feet^2、外れ値が2つあるようだ
* <YearBuilt>単位：年、1872〜2010の建設年のデータ

```python
# ndarrayに変換
X = train[['GrLivArea','YearBuilt']].values
y = train[['SalePrice']].values
```

week3のデータ分析で、SalePriceのデータ分布に偏りがあったので、  
今回も対数変換を行い、正規分布に近づける。

```python
# 対数変換
log_X = np.log1p(X)
log_y = np.log1p(y)
```

```python
# 変換後の散布図
fig,ax = plt.subplots(1,2,sharey=True,figsize=(16, 5))
ax[0].scatter(log_X[:,0],log_y)
ax[1].scatter(log_X[:,1],log_y);
```

```python
# 訓練データと、検証データに分割（75:25）
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(log_X, log_y, test_size=0.25, random_state=0)

print('データの分割結果の確認:')
print(' X_train:',X_train.shape)
print(' y_train:',y_train.shape)
print(' X_test :',X_test.shape)
print(' y_test :',y_test.shape)
```

```python
# 線形回帰
from sklearn.linear_model import LinearRegression

model = 'Linear'
clf = LinearRegression()

clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
# 評価
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

mse = mean_squared_error(np.expm1(y_test), np.expm1(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred))

evaluation_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

print('MSE : {:.3f}'.format(mse))
print('RMSE : {:.3f}'.format(rmse))
print('R2 : {:.3f}'.format(r2))
evaluation_df
```

```python
# 結果プロット
fig,ax = plt.subplots(figsize=(16, 9))

plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_test),label='test_data')
plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_pred),label='predict_data')

plt.title(model,size=20)
plt.xlabel('GrLivArea',size=16)
plt.ylabel('SalePrice',size=16)
plt.legend();
```

### 【問題3】手法の比較
以下の全ての手法について学習・推定・評価・可視化を行ってください。  
そして、指標値を１つの表にまとめてください。  
その表と可視化した図を利用して結果を簡単に考察してください。  
どの手法がどの指標値が良かったかや、どういった図になっていたかを述べてください。  
  
* 線形回帰
* SVM
* 決定木
* ランダムフォレスト

公式ドキュメントへのリンクは以下です。  
本来はパラメータチューニングを行いますが、まずはデフォルトの値で実行しましょう。  

* sklearn.linear_model.LinearRegression — scikit-learn 0.21.3 documentation

* sklearn.svm.SVR — scikit-learn 0.21.3 documentation

* sklearn.ensemble.RandomForestRegressor — scikit-learn 0.21.3 documentation

SVM、決定木、ランダムフォレストは分類問題でも使用した手法ですが、回帰問題用のライブラリはそれぞれ専用のものに分かれています。

```python
# SVM
from sklearn.svm import SVR

model = 'SVM'
clf = SVR(C=1.0, epsilon=0.2)

clf.fit(X_train, y_train.reshape(-1))
y_pred = clf.predict(X_test)
```

```python
# 評価

mse = mean_squared_error(np.expm1(y_test), np.expm1(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred))

tmp_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

evaluation_df = pd.concat([evaluation_df,tmp_df],axis=0)
```

```python
# 結果プロット
fig,ax = plt.subplots(figsize=(16, 9))

plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_test),label='test_data')
plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_pred),label='predict_data')

plt.title(model,size=20)
plt.xlabel('GrLivArea',size=16)
plt.ylabel('SalePrice',size=16)
plt.legend();
```

```python
# 決定木
from sklearn.tree import DecisionTreeRegressor

model = 'DecisionTree'
clf = DecisionTreeRegressor(random_state=0)

clf.fit(X_train, y_train.reshape(-1))
y_pred = clf.predict(X_test)
```

```python
# 評価
mse = mean_squared_error(np.expm1(y_test), np.expm1(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred))

tmp_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

evaluation_df = pd.concat([evaluation_df,tmp_df],axis=0)
```

```python
# 結果プロット
fig,ax = plt.subplots(figsize=(16, 9))

plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_test),label='test_data')
plt.scatter(np.expm1(X_test[:,0]),np.expm1(y_pred),label='predict_data')

plt.title(model,size=20)
plt.xlabel('GrLivArea',size=16)
plt.ylabel('SalePrice',size=16)
plt.legend();
```

```python
# ランダムフォレスト
from sklearn.ensemble import RandomForestRegressor

model = 'RandomForest'
clf = RandomForestRegressor(max_depth=2, random_state=0)

clf.fit(X_train, y_train.reshape(-1))
y_pred = clf.predict(X_test)
```

```python
# 評価
mse = mean_squared_error(np.expm1(y_test), np.expm1(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(np.expm1(y_test), np.expm1(y_pred))

tmp_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

evaluation_df = pd.concat([evaluation_df,tmp_df],axis=0)
```

### 【問題4】（アドバンス課題）他の特徴量も使用しての学習
ここまでは練習のために2つの特徴量だけを使いましたが、

### 検証１：標準化
アヤメの分類では、標準化を行い、スケールを合わせた。  
今回は、対数変換によるスケールを合わせた結果との比較をしてみる。

```python
# 説明変数と目的変数の抽出

features = ['GrLivArea','YearBuilt']

x = train[features].values
y = train[['SalePrice']].values
print('x:\n',x.shape)
print('y:\n',y.shape)
```

```python
# 訓練データと、検証データに分割（75:25）
X_train, X_test, y_train, y_test = \
train_test_split(x, y, test_size=0.25, random_state=0)

print('データの分割結果の確認:')
print(' X_train:',X_train.shape)
print(' y_train:',y_train.shape)
print(' X_test :',X_test.shape)
print(' y_test :',y_test.shape)
```

```python
# 標準化
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train_std = scaler.transform(X_train)
X_test_std = scaler.transform(X_test)

scaler.fit(y_train)
y_train_std = scaler.transform(y_train)
y_test_std = scaler.transform(y_test)
```

```python
# 標準化後のデータをプロットしてみる
fig,ax = plt.subplots(1,2,sharey=True,figsize=(16, 5))

ax[0].scatter(X_train_std[:,0],y_train_std)
ax[1].scatter(X_train_std[:,1],y_train_std);
```

```python
# SVM
model = 'SVM'
clf = SVR(C=1.0, epsilon=0.2)

clf.fit(X_train_std, y_train_std.reshape(-1))
y_pred = clf.predict(X_test_std)
```

```python
# 評価

mse = mean_squared_error(scaler.inverse_transform(y_test_std),
                         scaler.inverse_transform(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(scaler.inverse_transform(y_test_std),
              scaler.inverse_transform(y_pred))

evaluation_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

evaluation_df
```

```python
# 結果プロット
fig,ax = plt.subplots(figsize=(16, 9))

plt.scatter(X_test[:,0],scaler.inverse_transform(y_test_std),
            label='test_data')
plt.scatter(X_test[:,0],scaler.inverse_transform(y_pred),
            label='predict_data')

plt.title(model,size=20)
plt.xlabel('GrLivArea',size=16)
plt.ylabel('SalePrice',size=16)
plt.legend();
```

```python
# 決定木
model = 'DecisionTree'
clf = DecisionTreeRegressor(random_state=0)

clf.fit(X_train_std, y_train_std.reshape(-1))
y_pred = clf.predict(X_test_std)
```

```python
# 評価

mse = mean_squared_error(scaler.inverse_transform(y_test_std),
                         scaler.inverse_transform(y_pred))
rmse = np.sqrt(mse)
r2 = r2_score(scaler.inverse_transform(y_test_std),
              scaler.inverse_transform(y_pred))

evaluation_df = pd.DataFrame([[mse,rmse,r2]],
                             columns=['MSE','RMSE','R2'],index=[model])

evaluation_df
```

```python
# 結果プロット
fig,ax = plt.subplots(figsize=(16, 9))

plt.scatter(X_test[:,0],scaler.inverse_transform(y_test_std),
            label='test_data')
plt.scatter(X_test[:,0],scaler.inverse_transform(y_pred),
            label='predict_data')

plt.title(model,size=20)
plt.xlabel('GrLivArea',size=16)
plt.ylabel('SalePrice',size=16)
plt.legend();
```

### <考察>  
SVMは、対数変換をしたデータより、標準化したデータの方が結果（検証データの予測）が良くなった。  
決定木は、対数変換をしたデータの方が、僅かに指標値が良かった。

### 検証２：ランダムフォレスト
特徴量を、week3でSalePriceと相関関数が高かったTop10を使って、ランダムフォレストで推定する。
ここでランダムフォレストを採用した理由は、欠損値を処理せずに（理論については、今後、、、）  
学習できるため。

```python
# 説明変数と目的変数の抽出

features = ['OverallQual',
            'GrLivArea',
            'GarageCars',
            'GarageArea',
            'TotalBsmtSF',
            '1stFlrSF',
            'FullBath',
            'TotRmsAbvGrd',
            'YearBuilt',
            'YearRemodAdd']

x = train[features].values
y = train[['SalePrice']].values
print('x:\n',x.shape)
print('y:\n',y.shape)
```
