* Faster R-CNNの重要と考える点
1. RPN・・・resnet.py内のrpn関数で定義している。（入力は、baselayer、num_anchors）
2. RoIプーリング・・・RoIPoolingConv.pyで実装。（layerクラス）
3. Anchors Boxの設定・・・faster_rcnn.pyのget_model関数で定義し、rpn関数に渡せている。
4. Covlayerのfeature mapの共有・・・rpnとclassiferの両方がbaselayerを引数に取り、共有している。
5. 4step Alternating Training・・・losses.pyのrpn_loss_regr, rpn_loss-cls, class_loss_regr, class_loss_clsの4つの関数を定義している。iterごとにそれぞれを最適化し、4つの損失関数の合計が、全体のLossとしている。
