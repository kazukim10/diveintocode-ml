2.ダミーデータの作成

機械学習に関するプログラムの実験を行う上で、
実データではないランダムなデータ（ダミーデータ）が欲しいということがあります。
正規分布に従う乱数によるダミーデータの作成を通じてNumPyの使い方になれていきましょう。

【問題1】乱数の作成

```python
import numpy as np

np.random.seed(0)
mean0 = np.array([-3,0])
cov0 = np.array([[1, 0.8],
                [0.8, 1]])

data_0 = np.random.multivariate_normal(mean0,cov0,size = 500)
```

【問題2】散布図による可視化

```python
import matplotlib.pyplot as plt

x0 = data_0[:,0]
y0 = data_0[:,1]

plt.title("scatter")
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(x0,y0)
plt.show()
```

【問題3】ヒストグラムによる可視化

```python
plt.hist(x0,bins =50)
plt.title("histogram of x1")
plt.ylabel("frequency")
plt.xlabel("x1")
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.show()
```

```python
plt.hist(y0,bins =50)
plt.title("histogram of x2")
plt.ylabel("frequency")
plt.xlabel("x2")
plt.xlim(-5.5,2.5)
plt.ylim(0,30)
plt.show()
```

【問題4】データの追加

```python
np.random.seed(0)
mean1 = np.array([0,-3])
cov1 = np.array([[1, 0.8],
                [0.8, 1]])

data_1 = np.random.multivariate_normal(mean1,cov1,size = 500)
x1 = data_1[:,0]
y1 = data_1[:,1]
```

```python
plt.title("scatter")
plt.ylabel("x2")
plt.xlabel("x1")
plt.scatter(x0,y0,label = "0")
plt.scatter(x1,y1,label = "1")
plt.legend()
```

【問題5】データの結合

```python
data_concat = np.concatenate([data_0,data_1],axis = 0)
print(data_concat)
```

【問題6】ラベル付け

```python
data_0 = np.insert(data_0,2,0,axis = 1)
data_1 = np.insert(data_1,2,1,axis = 1)


print("data_0")
print(data_0)
print("data_1")
print(data_1)
```
